````python
#Q1
#! pip install tensorflow

from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np
df = mnist.load_data('minst.db')
train,test = df
X_train, y_train = train
X_test, y_test = test


np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5):
    print(y_test[i],'\n') # Print the label
    print(X_test[i],'\n') # Print the matrix of values
    plt.contourf(np.rot90(X_test[i].transpose())) # Make a contour plot of the matrix values
    plt.show()
````

````python
#Q2
print(X_train.shape)
print(X_test.shape)
print(X_train[0].shape)
print(X_test[0].shape)
print(y_train.shape)
print(y_test.shape)
````

X_train has 60000 matrixs with size of 28 by 28; X_test has 10000 matrixs with size of 28 by 28; each X_train and X_test value is a matrix of size 28 by 28; y_train has 60000 numerical assignemtns; y_test has 10000 numberical assignments

````python
#Q3
import pandas as pd
import numpy as np
import os

file_path = './data/Z_test.csv'
if os.path.exists(file_path):
    Z_train = pd.read_csv('./data/Z_train.csv')
    Z_test = pd.read_csv('./data/Z_test.csv')
else:
    Z_train = []
    for i in range(len(y_train)):
        row = X_train[i].reshape((1,784))
        Z_train.append(row[0])
    Z_train = pd.DataFrame(Z_train)
    Z_train.to_csv('./data/Z_train.csv')

    Z_test = []
    for i in range(len(y_test)):
        row = X_test[i].reshape((1,784))
        Z_test.append(row[0])
    Z_test = pd.DataFrame(Z_test)
    Z_test.to_csv('./data/Z_test.csv')
````

````python
#Q4
from sklearn.neighbors import KNeighborsClassifier
k_bar = 100
k_grid = np.arange(2,k_bar)
Acc = np.zeros(k_bar)
for k in range (k_bar):
    model = KNeighborsClassifier(n_neighbors=k+1)
    fitted_model = model.fit(Z_train.values,y_train)
    Acc[k] = model.score(Z_test.values,y_test)
Acc_max = np.max(Acc)
max_index = np.where(Acc == Acc_max)
k_star = max_index[0]+1
print (k_star)

plt.plot(np.arange(0,k_bar),Acc)
plt.xlabel("k")
plt.title("optimal k:"+str(k_star))
plt.ylabel('Accuracy')
plt.show()
````

````python
#Q5
knn = KNeighborsClassifier(n_neighbors=3)
model = knn.fit(Z_train.values,y_train)
y_hat = model.predict(Z_test.values)
Acc = model.score(Z_test.values,y_test)
print('Accuracy: ', Acc)
pd.crosstab(y_test, y_hat)
````

The model has a acurracy of 97% when k=3. It is clear that the model works quite well for distiguising different hand written numbers.
